```python
import numpy as np
  
def is_pos_def(x):
    return np.all(np.linalg.eigvals(x) > 0)

n = int(input("Enter the size of the matrix: "))

```

```python



is_good_enough = False
while not is_good_enough  :
    x = np.arange(1,n)
    V = []
    for i in range(n):
        y = np.random.randn(n)*2
        V.append(y)
    A = np.cov(V)
    is_good_enough = is_pos_def(A)

print ( is_pos_def(A))
print(A)
a = A
```

```python
%%time

l=np.linalg.cholesky(a) 
b=np.ones((n,1))
l1=np.linalg.inv(l) 
y=l1.dot(b) 
lt=np.transpose(l) 
lt1=np.linalg.inv(lt) 
x=lt1.dot(y) 
print(x)
```
